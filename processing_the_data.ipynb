```python
columns_to_convert = data.columns.difference(['Date', 'Time'])
data[columns_to_convert] = data[columns_to_convert].apply(pd.to_numeric, errors='coerce')
```

```python
# Columns to fill (excluding 'Date' and 'Time')
columns_to_fill = data.columns.difference(['Date', 'Time'])

# Fill missing values with the mean of each column (excluding 'Date' and 'Time')
data[columns_to_fill] = data[columns_to_fill].fillna(data[columns_to_fill].mean())
```

```python
data.isnull().sum()
```

```python
data.tail()
```

```python
#we forecast because time series data is given
```

```python
#Forecasting with FB Prophet Algorithm

#converting the date into format yyyy-mm-dd

date_info=pd.to_datetime(data['Date'],format="%d/%m/%Y")

print(date_info)
```

```python
time_info=data['Time']
```

```python
time_info=time_info.apply(lambda x:x.replace('.',':'))
print(time_info)
```

```python
#combining 2 series date_info and time_info into a single dataframe
date_time=pd.concat([date_info,time_info],axis=1)
date_time.head()
```

```python
#combining Date and time column to form a single ds column, (prophet algo requires 2 columns ds and y)

date_time['ds']=date_time['Date'].astype(str)+' '+date_time['Time'].astype(str)
```

```python
date_time.head()
```

```python
date_time.info()
```

```python
#creating a new dataframe that only contains ds and y(any column to forecast)
data1=pd.DataFrame()
data1['ds']=pd.to_datetime(date_time['ds'])
```

```python
data1.head()
```

```python
data1.info()
```

```python
data1['y']=data['RH']
```

```python
data1.to_csv('datetimestamp_RH.csv', index=False)
```

```python
from prophet import Prophet
```

```python
#training the prophet model
model=Prophet()
model.fit(data1)
```

```python
#lets forecast future value of RH
future=model.make_future_dataframe(periods=365,freq='H')
future.tail()
```

```python
forecast=model.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```
